# Q2: Make dash that takes name input and outpt "Welcome to Data Science dashboard {name}"

```python
app.layout = html.Div([
    dcc.Input(id='input',value='',),
    html.Div(id='output', style={'whiteSpace': 'pre-line'})
])

@callback(
    Output('output', 'children'),
    Input('input', 'value')
)
def update_output(value):
    return 'Welcome to Data Science dashboard {}'.format(value)

app.run_server(port=9877)
```

# Q7: From our data make plotly-dash with suitble chart, and Dropdown.

- Dropdown should contain at least 3 options, each one represents different column.
- Add label to dropdown.
- The chart data should change depending on the selected dropdown option.

```python
import pandas as pd
df = pd.read_csv("titanic passenger list.csv")
df["survived"] = df["survived"].astype(str)
df["survived"][df["survived"] == "1.0"] = "alive"
df["survived"][df["survived"] == "0.0"] = "died"
df
```

```python
dfxx = pd.DataFrame({"sex": ["male", "female"],
                        "count": [841, 464]})

dfxx
```

```python
app = JupyterDash()



app.layout = html.Div(children=[
    html.Br(),
    html.H1('Survived Counts', style={'text-align': 'center', 'color': 'white'}),

    # Dropdown
    dcc.Dropdown(
        id='my-drop',
        options=["age", "sex", "survived"],
        value='age' # Set initial value
    ),

    # Graph
    dcc.Graph(id='my-graph')
])


@app.callback(
    Output('my-graph', 'figure'),
    Input('my-drop', 'value')
)
def update_graph(val):
    
    filtered_df = df[val]
    bar_chart = px.histogram(filtered_df, x=val, title=f'Survived in {val} ')
    return bar_chart


if __name__ == '__main__':
    app.run()
```
